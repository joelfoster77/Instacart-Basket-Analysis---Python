```python
#import pandas, numpy, and os libraries
import pandas as pd
import numpy as np
import os
path = r'C:\Users\joelf\OneDrive\Documents\CareerFoundry\4 Python\Instacart Data'
```

```python
#1
#import ords_prior
ords_prods_merged = pd.read_pickle(os.path.join(path, 'NewDataSets', 'orders_products_combined.pkl'))
ords_prods_merged.shape
```

```python
#2
#Find mean of order_number column when grouped by department_id
ords_prods_merged.groupby('department_id')['order_number'].mean()
```

#3
For the most part the means for the whole dataset are lower. The dataset of the first 1000000 rows had users with more orders on avg
I would speculate that lower user_id numbers (included in the first 1m rows) are assigned to the oldest instacart members and older members have more time to place orders. Including the whole dataset brings in newer users with fewer orders.

```python
#4a
#create max order column and pop with max number of orders for that customer to see who has the most orders
ords_prods_merged['max_order'] = ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
ords_prods_merged.head(15)
```

```python
#4b
#loyalty flags for whole dataset
ords_prods_merged.loc[ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal Customer'
ords_prods_merged.loc[ (ords_prods_merged['max_order'] <= 40) & (ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular Customer'
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New Customer'
ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
#4c
#checking to make sure flags worked
ords_prods_merged[['user_id','loyalty_flag','max_order']].head(60)
```

```python
#5
#find average price of products when grouped by loyalty flag
ords_prods_merged.groupby('loyalty_flag')['prices'].mean()
```

Average product price is lowest for loyal customers

```python
#6
#create column with avg price for each customer
#spending flag based on avg price across orders
#   mean < 10 = low spender
#   mean >=10 = high spender
ords_prods_merged['avg_price'] = ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
ords_prods_merged.head(15)
```

```python
#6 cont.
ords_prods_merged.loc[ords_prods_merged['avg_price'] < 10, 'spender_flag'] = 'Low Spender'
ords_prods_merged.loc[ords_prods_merged['avg_price'] >= 10, 'spender_flag'] = 'High Spender'
ords_prods_merged['spender_flag'].value_counts(dropna = False)
```

```python
ords_prods_merged.head(50)
```

```python
#7
#create column with median days since prior order
#frequency flag based on median time since prior order
#   median > 20 = non frequent customers
#   median > 10 and <= 20 = regular customer
#   median <= 10 = frequent customer
ords_prods_merged['med_order_freq'] = ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
ords_prods_merged.head(15)
```

```python
#7 cont.
ords_prods_merged.loc[ords_prods_merged['med_order_freq'] > 20, 'freq_flag'] = 'Nonfrequent Customer'
ords_prods_merged.loc[ (ords_prods_merged['med_order_freq'] <=20) & (ords_prods_merged['med_order_freq'] > 10), 'freq_flag'] = 'Regular Customer'
ords_prods_merged.loc[ords_prods_merged['med_order_freq'] <= 10, 'freq_flag'] = 'Frequent Customer'
ords_prods_merged['freq_flag'].value_counts(dropna = False)
```
